```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
df = df.dropna(subset=['date_added'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

**Insight:**  
Movies are more common on Netflix compared to TV Shows, indicating a stronger focus on movie content.


```python
df['release_year'].value_counts().sort_index().plot(figsize=(10,5))
plt.title("Content Release Trend Over Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

**Insight:**  
Netflix content production has increased significantly after 2015, showing rapid platform growth.


```python
df['country'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Countries by Content Count")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.show()
```

**Insight:**  
The United States contributes the highest number of titles on Netflix, followed by other major content-producing countries.


```python
df['listed_in'].str.split(',').explode().value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

**Insight:**  
Drama and Comedy are the most frequently appearing genres on Netflix.


```python
movies = df[df['type'] == 'Movie']
movies['duration_minutes'] = movies['duration'].str.replace(' min', '').astype(float)

sns.histplot(movies['duration_minutes'], bins=30)
plt.title("Movie Duration Distribution")
plt.xlabel("Duration (minutes)")
plt.show()
```

**Insight:**  
Most Netflix movies fall within a common duration range, suggesting standardized content lengths.


### Key Insights
- Netflix has rapidly expanded its content library after 2015.
- Movies dominate the platform over TV Shows.
- The United States is the largest content contributor.
- Drama and Comedy are the most popular genres.
- Movie durations follow a consistent pattern.

### Conclusion
This exploratory data analysis highlights important trends in Netflix’s content strategy and demonstrates how data analysis can support business and content decisions.
